```python
#imorting required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
#reading the csv file
df=pd.read_csv(r"E:\Education_data_analysis_1\Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#impact of parent education on students
gb=df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.show()
```
