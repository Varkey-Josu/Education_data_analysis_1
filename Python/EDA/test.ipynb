```python
#imorting required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the csv file
df=pd.read_csv(r"E:\Education_data_analysis_1\Expanded_data_with_more_features.csv")
df.head()
```

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#impact of parent education on students
gb=df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("Parent Education vs Students Score")
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Parents Maritial Status vs Students Score")
plt.show()
```

```python
sns.boxplot(data=df, x="MathScore")
plt.title("Maths Score")
plt.show()
```

```python
sns.boxplot(data=df, x="WritingScore")
plt.title("Writing Score")
plt.show()
```

```python
sns.boxplot(data=df, x="ReadingScore")
plt.title("Reading Score")
plt.show()
```

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()

l=["groupA","groupB","groupC","groupD","groupE"]
mlst=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.title("Distribution of Ethnic Group")
plt.pie(mlst, labels=l, autopct="%1.2f%%")
```
